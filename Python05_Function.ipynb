```python
def sum_list(numbers):
    if len(numbers) == 0:
        return 0 
    else:
        return numbers[0] + sum_list(numbers[1:])
my_list = [1, 2, 3, 4, 5]
print(sum_list(my_list))
```

```python
def find_max(numbers):
    if len(numbers) == 1:
        return numbers[0]
    else:
        max_of_rest = find_max(numbers[1:])
        return numbers[0] if numbers[0] > max_of_rest else max_of_rest
my_list = [3, 7, 2, 9, 1]
print(find_max(my_list))
```

```python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1
for num in count_up_to(5):
    print(num)    
```

```python
def large_sequence(n):
    for i in range(n):
        yield i
# This doesn't create a million numbers in memory
gen = large_sequence(1000000)
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
def fibonacci():
    a, b = 0, 1
    while True: 
        yield a
        a, b = b, a + b
# Get first 100 Fibonacci numbers
gen = fibonacci()
for _ in range(100):
    print(next(gen))
```

```python
def echo_generator():
    while True:
        received = yield
        print("Received:", received)
gen = echo_generator()
next(gen) # Prime the generator
gen.send("Hello")
gen.send("World")
```

```python
def my_gen():
    try:
        yield 1
        yield 2
        yield 3
    finally:
        print("Generator closed")
gen = my_gen()
print(next(gen))
gen.close()
```
